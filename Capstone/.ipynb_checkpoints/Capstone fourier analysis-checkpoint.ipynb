```python
import numpy as np
import pandas as pd
import scipy.optimize as opt
from scipy.interpolate import interp1d
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.misc
from scipy.fft import fft, fftfreq
```

```python
'''We will have intensity data I and time data t from logger pro
We will need to calculate path length change data ds using angular velocity (or similar) of servo motor'''

#data

raw_data_array = pd.read_csv('test.csv')
raw_data = pd.DataFrame(raw_data_array).to_numpy()
I = raw_data[:,1]
t = raw_data[:,0]
w = 1  #not true. just choosing a value so i can proceed for now
ds = (50/(2*np.pi))*w*t    #for angular velocity w in radians

#graph data pre transformation

fig, ax = plt.subplots(figsize=(10,6))
ax.plot(ds,I)
plt.ylabel("Intensity of center of pattern (lux)")
plt.xlabel("Path length difference (m)")
plt.title("Intensity of center of diffraction pattern with varying path length difference")
plt.show()
```

```python
#Fourier transform

n = len(I)

yf = fft(I)
xffreq = fftfreq(n, 1/1000) #will have to change this to in terms of dt later
xfwavelength = 2*np.pi/xffreq

plt.plot(xffreq, np.abs(yf))
plt.xlim(-50,50)
plt.show()
```
