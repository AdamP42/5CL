```python
# Does the fit
init_params=[0, 0]
lin_opt, lin_cov = opt.curve_fit(model, freq, energy, xerr=freq_err, yerr=energy_err, absolute_sigma=True, p0=init_params)
plt.title("Exp 3: Rough Analysis of Planks Constant and Cuttoff Freq, \n")

# Best fit values
m, b = lin_opt
dm, db = np.sqrt(np.diag(lin_cov))
m_unc = ufloat(m, dm)
b_unc = ufloat(b, db)

plt.errorbar(freq, energy, xerr=freq_err, yerr=energy_err, fmt='o')
plt.plot(freq, model(freq, m, b), color='orange')
plt.ylabel("Energy (eV)")
plt.xlabel("Frequency (Hz)")

plt.show()
```
